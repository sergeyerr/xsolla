```python
from __future__ import print_function
import time
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_mldata
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import random
```

```python
segment_directory = "data/segmentation.csv"

segment_init = pd.read_csv(segment_directory, header=0, parse_dates = [2] )
segment = segment_init.copy()
segment['Recency'] = ((segment['Recency'].max() - segment['Recency']) / np.timedelta64(1, 'D')).astype(int) #дни с последней закупки
```

```python
segment[['factor_1']].plot.hist() #люблю запах равномерных распределений по утрам
segment[['factor_2']].plot.hist()
segment[['factor_3']].plot.hist()
segment[['factor_4']].plot.hist()
```

```python
segment['Monetary'].hist()
```

```python
segment['Monetary'] = np.floor(segment['Monetary'] / 100)
```

```python
segment['Monetary'].where(segment['Monetary'] < 2, 2, inplace = True) # разделим monetary на 3 части 0-100, 100-200, 200+
segment['Monetary'].hist() # в последствии алгоритмом будем делить на 3 кластера, надеясь на то, что за каждым кластером будет
                           #своя категория monetary
```

```python
#уменьшим размерность, далее запихаем в k-means вместе с monetary
pca = PCA(n_components=3) 
pca_result = pca.fit_transform(segment[['Recency','Frequency','factor_1', 'factor_2' ,'factor_3' , 'factor_4']].values)
components_relation = pd.DataFrame(pca.components_,columns=['Recency','Frequency','factor_1', 'factor_2' ,'factor_3' , 'factor_4'],index = ['pca-one','pca-two', 'pca-three'])
segment['pca-one'] = pca_result[:,0]
segment['pca-two'] = pca_result[:,1] 
segment['pca-three'] = pca_result[:,2] 
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
print('components relations with features:')
print(components_relation)
```

```python
segment.corr()['Monetary'] # как видно, factor_* не имеют влияения на monetary, уберём их далее 
```

```python
rndperm = list(range(1,1000))

ax = plt.figure(figsize=(16,10)).gca(projection='3d')
ax.scatter(
    xs=segment.loc[rndperm,:]["Frequency"], 
    ys=segment.loc[rndperm,:]["Recency"],
    zs=segment.loc[rndperm,:]["Monetary"],
    
    cmap='tab10'
)
ax.set_xlabel("Frequency")
ax.set_ylabel("Recency")
ax.set_zlabel('monetary')
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np
reduced = segment[['Recency','Frequency','Monetary']][:]
for col in ['Recency','Frequency', 'Monetary']:
    reduced[col] = (reduced[col] - reduced[col].min())/(reduced[col].max()-reduced[col].min())
kmeans = KMeans(n_clusters=3, random_state=0).fit(reduced)
reduced['label'] = kmeans.labels_
#попробуем k-means, посмотрим на результаты
```

```python
for i in range(3):
    print('Статистика по кластеру ', i)
    print(segment_init[reduced['label'] == i].describe())
```

```python
rndperm = list(range(1,1000))
for_draw = segment_init.iloc[rndperm]
for_draw['label'] = reduced['label'].copy()
for kl in range(3):
    plt.scatter(for_draw.loc[for_draw['label'] == kl,'Frequency'],for_draw.loc[for_draw['label'] == kl, 'Recency'], label = kl, alpha=0.1)
plt.xlabel('Frequency')
plt.ylabel('Date')
plt.legend()
plt.show()
```
